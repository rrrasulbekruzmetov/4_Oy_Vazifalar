```python
import matplotlib.pyplot as plt # malumotlarni chizish (vizualizatsiya) uchun kutubxona.
import seaborn as sns # malumotlarni chizish (vizualizatsiya) uchun kutubxona.
from sklearn.datasets import make_blobs # klasterlash uchun suniy malumotlar (nuqtalar) hosil qiladi.
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster # hierarchical clustering (ierarxik klasterlash) amallari uchun kutubxona 

# 1. Namunaviy malumotlarni yaratish
X, _ = make_blobs(n_samples=10, centers=3, random_state=42) # 10 ta nuqta yaratiladi. //  10 ta nuqta yaratiladi. // natija doim bir xil chiqishi uchun random  berilgan qiymat. // x-nuqtalar koordinatalari.
 

Z = linkage(X, method='ward')  # Ward usuli klaster ichidagi farqlarni (dispersiyani) eng kichik qiladi. // z- linkage matritsasi (qaysi nuqta qaysi bilan birlashtirilganini va masofani korsatadi).
 

plt.figure(figsize=(8, 4)) ## grafik olchamini belgilaydi. 
dendrogram(Z) # ierarxik daraxt (dendrogramma) ni chizadi.
plt.title("Hierarchical Clustering Dendrogram") # grafikning sarlavhasi va oqlarini belgilaydi.
plt.xlabel("Data Points") # grafikning sarlavhasi va oqlarini belgilaydi.
plt.ylabel("Distance") # grafikning sarlavhasi va oqlarini belgilaydi.
plt.show()  # grafikni ekranga chiqaradi.


clusters = fcluster(Z, t=3, criterion='maxclust') # linkage natijalaridan foydalanib, 3 ta klaster boyicha nuqtalarni ajratadi. // t - maksimal klasterlar soni 3 bolsin. //  cri_ klasterlarni maksimal son boyicha ajratish mezoni.
print("Cluster labels:", clusters) # natijani ekranga chiqaradi.
```
